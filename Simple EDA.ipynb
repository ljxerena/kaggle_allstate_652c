```python
catvar=[col for col in train_data.columns if 'cat' in col]
```

```python
print('categorical variables ',catvar)
```

```python
for col in catvar[109:115]:
    sns.countplot(train_data[col])
    sns.plt.title(col)
    plt.show()
```

```python
#unique value count in each catgorial variable
for col in catvar:
    print(train_data[col].value_counts())
```
